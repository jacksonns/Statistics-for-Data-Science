## Quest√£o 2



```python
from scipy import stats
import matplotlib.pyplot as plt
```

```python
X = []
Y = []
for _ in range(1000):
    x = stats.uniform.rvs(0, 10)
    y = stats.uniform.rvs(x/3, 2*x)
    X.append(x)
    Y.append(y)
```

Os valores poss√≠veis de Y s√£o o intervalo [0,20] 

```python
plt.scatter(X, Y)
plt.show()
```

Os valores de Y n√£o parecem seguir uma distribui√ß√£o Normal, a densidade est√° mais concentrada pr√≥xima de 0.

```python
plt.hist(Y, bins=15)
plt.show()
```

Esperan√ßa estimada = 6.98

Varian√ßa estimada = 27.47

```python
Esp = sum(Y) / len(Y)
Esp
```

```python
soma_quadrados = sum((y - Esp) ** 2 for y in Y)
Var = soma_quadrados / (len(Y) -1 )
Var
```

## Quest√£o 5



Pela lei da Vari√¢ncia Total, $V(Y) = E[V(Y|X)] + V(E[Y|X])$, por essa f√≥rmula, obtemos 

$V(Y) - E[V(Y|X)] = V(E[Y|X])$

O termo √† esquerda $V(Y) - E[V(Y|X)]$ √© n√£o negativo, pois representa a diferen√ßa entre a vari√¢ncia total de Y e a expectativa da vari√¢ncia condicional de Y dado X. A vari√¢ncia n√£o pode ser negativa, e a esperan√ßa de uma quantidade n√£o negativa tamb√©m √© n√£o negativa.

Portanto, temos $V(Y) - E[V(Y|X)] \geq 0$. Ou seja, $V(Y) \geq E[V(Y|X)]$
